## Activation functions

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.special import expit as sigmoid
```

```python
def relu(z):
    return np.maximum(0, z)

def derivative(f, z, eps=0.000001):
    return (f(z + eps) - f(z - eps)) / (2 * eps)

def derivative_relu(z):
    return (z >= 0).astype(z.dtype)

z = np.linspace(-5, 5, 200)
plt.style.use("seaborn-v0_8-notebook")

plt.figure(figsize=(14, 12))
plt.subplots_adjust(hspace=0.4)

# Heaviside
plt.subplot(4, 2, 1)
plt.plot([-5, 0], [0, 0], color='#1976d2', linestyle='-', linewidth=2)
plt.plot([0, 0], [0, 1], color='#1976d2', linestyle='--', linewidth=2)
plt.plot([0, 5], [1, 1], color='#1976d2', linestyle='-', linewidth=2)
plt.title("Heaviside", fontsize=14)

plt.subplot(4, 2, 2)
plt.plot(z, derivative(np.sign, z), color='#ef5350', linestyle='-', linewidth=2)
plt.title("Derivative of Heaviside", fontsize=14)

# Sigmoid
plt.subplot(4, 2, 3)
plt.plot(z, sigmoid(z), color='#1976d2', linestyle='-', linewidth=2)
plt.title("Sigmoid", fontsize=14)

plt.subplot(4, 2, 4)
plt.plot(z, derivative(sigmoid, z), color='#ef5350', linestyle='-', linewidth=2)
plt.title("Derivative of Sigmoid", fontsize=14)

# Hyperbolic Tangent
plt.subplot(4, 2, 5)
plt.plot(z, np.tanh(z), color='#1976d2', linestyle='-', linewidth=2)
plt.title("Hyperbolic Tangent", fontsize=14)

plt.subplot(4, 2, 6)
plt.plot(z, derivative(np.tanh, z), color='#ef5350', linestyle='-', linewidth=2)
plt.title("Derivative of Hyperbolic Tangent", fontsize=14)

# ReLU
plt.subplot(4, 2, 7)
plt.plot(z, relu(z), color='#1976d2', linestyle='-', linewidth=2)
plt.title("ReLU", fontsize=14)

plt.subplot(4, 2, 8)
plt.plot(z, derivative_relu(z), color='#ef5350', linestyle='-', linewidth=2)
plt.title("Derivative of ReLU", fontsize=14)
```
